```python
# Importa a biblioteca requests para fazer requisições HTTP à API da Riot
import requests
import time

API_KEY = 'RGAPI-ce5edec1-c09a-45ee-8619-68e01e462780'
headers = {"X-Riot-Token": API_KEY}

# 1. Obter os jogadores do Challenger (Coreia)
# Requisição para obter os jogadores do elo Challenger no modo ranqueado solo 5v5 do servidor coreano
challenger_url = "https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5"
response = requests.get(challenger_url, headers=headers)
if response.status_code != 200:
    print("Erro ao obter dados Challenger:", response.status_code, response.text)
    exit()

data_challenger = response.json()
players = data_challenger.get('entries', [])
```

```python
top_players = players[:50];top_players
```

```python
summoner_ids = [player["summonerId"] for player in top_players];summoner_ids
```

```python
puuids = []
for summoner_id in summoner_ids:
    summoner_url = f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/{summoner_id}"
    response = requests.get(summoner_url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        puuids.append(data.get("puuid"))
    else:
        print(f"Erro ao obter dados do summoner {summoner_id}: {response.status_code}, {response.text}")
    time.sleep(1.2)  

print("PUUIDs obtidos:", puuids)
```

```python
match_ids = set()  
for puuid in puuids:
# Recupera os IDs das 10 partidas mais recentes de cada jogador (via PUUID)
    match_history_url = f"https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count=20"
    response = requests.get(match_history_url, headers=headers)
    if response.status_code == 200:
        ids = response.json()  # lista de match IDs
        match_ids.update(ids)
    else:
        print(f"Erro ao obter partidas para puuid {puuid}:", response.status_code, response.text)
    time.sleep(1.2)  

match_ids = list(match_ids)
print(f"Foram coletados {len(match_ids)} match IDs.")
```

```python
match_ids
```

```python
matches_data = []
max_retries = 5  # Número máximo de tentativas por partida

# Limitar a apenas 600 partidas
limited_match_ids = match_ids[:600]

for match_id in limited_match_ids:
    match_url = f"https://asia.api.riotgames.com/lol/match/v5/matches/{match_id}"
    retries = 0  # Contador de tentativas

    while retries < max_retries:
        response = requests.get(match_url, headers=headers)

        if response.status_code == 200:
            match_data = response.json()
            matches_data.append(match_data)
            break  # Sai do loop se a requisição foi bem-sucedida
        
        elif response.status_code == 429:
            wait_time = 10 * (2 ** retries)  # Espera cresce exponencialmente: 10s, 20s, 40s...
            print(f"⚠️ Rate limit atingido! Esperando {wait_time} segundos...")
            time.sleep(wait_time)
            retries += 1  # Conta a tentativa
        
        else:
            print(f"Erro ao obter partida {match_id}: {response.status_code} - {response.text}")
            break  # Sai do loop em outros erros

    time.sleep(1.2)  # Delay padrão entre requisições

print(f"✅ Coleta finalizada! Total de partidas coletadas: {len(matches_data)}")
```

```python
queue_ids = set()  # Usando um set para garantir que os valores sejam únicos

for match in matches_data:
    try:
        queue_id = match['info'].get('queueId')
        if queue_id is not None:
            queue_ids.add(queue_id)  # Adiciona o queueId ao conjunto
    except Exception as e:
        continue  # Ignora erros e continua o loop

# Imprimir todos os queueIds únicos encontrados
print(f"QueueIds encontrados: {queue_ids}")
```

```python
#Código retificado

rows = []
for match in matches_data:
    try:
        match_id = match['metadata']['matchId']
        info = match.get('info', {})
        game_duration = info.get('gameDuration')  # Target (em segundos)
        game_mode = info.get('gameMode', None)
        
            
        blue_kills = 0
        red_kills = 0
        blue_gold = 0
        red_gold = 0
        blue_deaths = 0
        red_deaths = 0
        blue_assists = 0
        red_assists = 0
        blue_firstTower = 0
        red_firstTower = 0
        
        for participant in info.get('participants', []):
            team = participant.get('teamId')
            if team == 100:
                blue_kills += participant.get('kills', 0)
                blue_deaths += participant.get('deaths', 0)
                blue_assists += participant.get('assists', 0)
                blue_gold += participant.get('goldEarned', 0)
                if participant.get('firstTower'):
                    blue_firstTower += 1
            elif team == 200:
                red_kills += participant.get('kills', 0)
                red_deaths += participant.get('deaths', 0)
                red_assists += participant.get('assists', 0)
                red_gold += participant.get('goldEarned', 0)
                if participant.get('firstTower'):
                    red_firstTower += 1
        
        gold_diff = blue_gold - red_gold
        winner = 1 if blue_gold > red_gold else 0
        
        # Cálculo do KDA
        blue_kda = round((blue_kills + blue_assists) / max(1, blue_deaths), 2)
        red_kda = round((red_kills + red_assists) / max(1, red_deaths), 2)
        
        # Diferença de ouro por minuto
        gold_diff_per_min = round(gold_diff / max(1, (game_duration / 60)), 2)
        
        teams = info.get('teams', [])
        if len(teams) >= 2:
            team_blue = None
            team_red = None
            for t in teams:
                if t.get('teamId') == 100:
                    team_blue = t
                elif t.get('teamId') == 200:
                    team_red = t
            if team_blue and team_red:
                blue_tower_kills = team_blue.get('objectives', {}).get('tower', {}).get('kills', 0)
                red_tower_kills = team_red.get('objectives', {}).get('tower', {}).get('kills', 0)
                blue_dragon_kills = team_blue.get('objectives', {}).get('dragon', {}).get('kills', 0)
                red_dragon_kills = team_red.get('objectives', {}).get('dragon', {}).get('kills', 0)
                blue_baron_kills = team_blue.get('objectives', {}).get('baron', {}).get('kills', 0)
                red_baron_kills = team_red.get('objectives', {}).get('baron', {}).get('kills', 0)
                blue_inhibitor_kills = team_blue.get('objectives', {}).get('inhibitor', {}).get('kills', 0)
                red_inhibitor_kills = team_red.get('objectives', {}).get('inhibitor', {}).get('kills', 0)
                blue_riftHerald_kills = team_blue.get('objectives', {}).get('riftHerald', {}).get('kills', 0)
                red_riftHerald_kills = team_red.get('objectives', {}).get('riftHerald', {}).get('kills', 0)
            else:
                blue_tower_kills = red_tower_kills = 0
                blue_dragon_kills = red_dragon_kills = 0
                blue_baron_kills = red_baron_kills = 0
                blue_inhibitor_kills = red_inhibitor_kills = 0
                blue_riftHerald_kills = red_riftHerald_kills = 0
        else:
            blue_tower_kills = red_tower_kills = 0
            blue_dragon_kills = red_dragon_kills = 0
            blue_baron_kills = red_baron_kills = 0
            blue_inhibitor_kills = red_inhibitor_kills = 0
            blue_riftHerald_kills = red_riftHerald_kills = 0
        
        tower_diff = blue_tower_kills - red_tower_kills
        dragon_diff = blue_dragon_kills - red_dragon_kills
        baron_diff = blue_baron_kills - red_baron_kills
        inhibitor_diff = blue_inhibitor_kills - red_inhibitor_kills
        riftHerald_diff = blue_riftHerald_kills - red_riftHerald_kills
        
        row = {
            'matchId': match_id,
            'gameDuration': game_duration,
            'gameMode': game_mode,
            'winner': winner,
            'blue_kills': blue_kills,
            'red_kills': red_kills,
            'blue_gold': blue_gold,
            'red_gold': red_gold,
            'gold_diff': gold_diff,
            'gold_diff_per_min': gold_diff_per_min,
            'blue_kda': blue_kda,
            'red_kda': red_kda,
            'blue_firstTower_count': blue_firstTower,
            'red_firstTower_count': red_firstTower,
            'blue_tower_kills': blue_tower_kills,
            'red_tower_kills': red_tower_kills,
            'tower_diff': tower_diff,
            'blue_dragon_kills': blue_dragon_kills,
            'red_dragon_kills': red_dragon_kills,
            'dragon_diff': dragon_diff,
            'blue_baron_kills': blue_baron_kills,
            'red_baron_kills': red_baron_kills,
            'baron_diff': baron_diff,
            'blue_inhibitor_kills': blue_inhibitor_kills,
            'red_inhibitor_kills': red_inhibitor_kills,
            'inhibitor_diff': inhibitor_diff,
            'blue_riftHerald_kills': blue_riftHerald_kills,
            'red_riftHerald_kills': red_riftHerald_kills,
            'riftHerald_diff': riftHerald_diff,
        }
        rows.append(row)
    except Exception as e:
        print(f"Erro ao processar match {match.get('metadata', {}).get('matchId', 'N/A')}: {e}")

# Importa a biblioteca pandas para manipulação de dados tabulares
import pandas as pd

# Constrói um DataFrame pandas com os dados coletados
df = pd.DataFrame(rows)
print("Exemplo de registro extraído:")
print(df.head())

# Exporta os dados para um arquivo CSV
df.to_csv('regression400_dataset.csv', index=False)
print("Dataset de regressão exportado para 'regression_dataset.csv'.")
```
